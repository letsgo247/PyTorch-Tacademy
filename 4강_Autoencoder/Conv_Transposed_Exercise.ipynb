```python
import torch
import torch.nn as nn
import torch.nn.init as init
```

```python
img = torch.ones(1,1,3,3)
print(img)
```

```python
transpose1 = nn.ConvTranspose2d(in_channels=1, out_channels=1, kernel_size=3, 
                               stride=1, padding=0, output_padding=0, bias=False)
```

```python
print(transpose1.weight.data)
```

```python
init.constant_(transpose1.weight.data, 1)
```

```python
transpose1(img)
```

```python
transpose2 = nn.ConvTranspose2d(in_channels=1, out_channels=1, kernel_size=3, 
                               stride=2, padding=0, output_padding=0, bias=False)
init.constant_(transpose2.weight.data, 1)
transpose2(img)
```

```python
transpose3 = nn.ConvTranspose2d(in_channels=1, out_channels=1, kernel_size=3, 
                               stride=2, padding=1, output_padding=0, bias=False)
init.constant_(transpose3.weight.data, 1)
transpose3(img)
```

```python
transpose4 = nn.ConvTranspose2d(in_channels=1, out_channels=1, kernel_size=3, 
                               stride=2, padding=0, output_padding=1, bias=False)
init.constant_(transpose4.weight.data, 1)
transpose4(img)
```

```python
transpose5 = nn.ConvTranspose2d(in_channels=1, out_channels=1, kernel_size=3, 
                               stride=2, padding=1, output_padding=1, bias=False)
# output_padding 먼저 갈기고, 그 다음 padding으로 후리는듯?
init.constant_(transpose5.weight.data, 1)
transpose5(img)
```
