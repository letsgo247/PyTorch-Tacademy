```python
x = Variable(torch.FloatTensor(3,4), requires_grad=True)
y = x**2 + 4*x
z = 2*y + 3
```

```python
x.requires_grad, y.requires_grad, z.requires_grad
```

```python
loss = torch.FloatTensor(3,4)
z.backward(loss)

print(x.grad)
y.grad, z.grad
```
